```python
import pandas as pd
import matplotlib.pyplot as plt
```

Notebook om snel oude en nieuwe resultaten te vergelijken, waarbij de oude methode geen opsplitsing per jaar doet en de nieuwe wel

```python
df_new = pd.read_csv('../outputs/output_analysis/bare_soil_analysis_2022_v3.csv', index_col = 0)

pointids = ['Lm_11070','Lm_11568.1','Lm_15064.9', 'Lm_8486.3', 'Lm_8898.1.8', 'Lm_9709.1.6', 'Zn_10111.7', 'Zn_10583.5', 'Zn_11411.7']
df_old = df_new.loc[(df_new.year == 2022) & (df_new.pointid.isin(pointids))]


# Create a pivot table to restructure the data for grouped bar plotting
pivot_data = df_old.pivot(index='pointid', columns='period', values='% bare soil')

# Plot a grouped bar chart
pivot_data.plot(kind='bar', figsize=(12, 6))
plt.title('Percentage across periods for each pointid')
plt.xlabel('Code')
plt.ylabel('Percentage')
plt.legend(title='Period')
plt.xticks(rotation=45)
plt.show()
```
